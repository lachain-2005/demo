## 5. Armstrong Filter

Function `find_armstrong(nums, power=3)`.

### Sample Input:
```
find_armstrong([153, 370, 100, 9474])
```

### Sample Output:
```
[153, 370]
```

````python
# Write your solution here
def find_armstrong(nums,power=3):
  armstrong=[]
  for i in nums:
    sum=0
    x=str(i)
    for j in x:
      sum+=int(j)**power
    if sum==i:
     armstrong.append(i)
  print(armstrong)
nums=[153,255,370,100]
find_armstrong(nums)

````

## 9. List Statistics

### Sample Input:
```
stats([2,5,8,11,14])
```

### Sample Output:
```
Max: 14
Min: 2
Average: 8
Even: 3
Odd: 2
```

````python
# Write your solution here
def stats(x):
  max=0
  min=x[0]
  even_count=0
  odd_count=0
  count=0
  sum=0
  for i in x:
    if i>max:
      max=i
    elif i<min:
      min=i
    count+=1
    sum+=i
    if i%2==0:
      even_count+=1
    else:
      odd_count+=1
  average=sum/count
  print(f"max={max}\nmin={min}\naverage={average}\neven_count={even_count}\nodd_count={odd_count}")

x=[2,5,8,11,14]
stats(x)


````

## 10. Palindrome Checker

### Sample Input:
```
find_palindromes(["level","cat","racecar"])
```

### Sample Output:
```
['level','racecar']
```

````python
# Write your solution here
def find_palindrome(x):
  palindrome=[]
  for words in x:
   if words==words[::-1]:
    palindrome.append(words)
  print(palindrome)
x=["level","cat","racecar"]
find_palindrome(x)
````

## 11. Hollow Square

```

*   *
```

````python
# Write your solution here
def hollow_square(a):
  print("*"*a)
  for i in range (a-2):
    print("*"+" "*(a-2)+"*")
  print("*"*a)

hollow_square(5)
````

## 15. Frequency Counter

### Sample Input:
```
clean_and_count("Hello World")
```

### Sample Output:
```
{'h':1,'e':1,'l':3,'o':2,'w':1,'r':1,'d':1}
```

````python
# Write your solution here
def frequency(x):
  x=x.replace(" ","").lower()
  freq={}
  for i in x:
    if i in freq:
      freq[i]+=1
    else:
      freq[i]=1
  print(freq)
frequency("Hello world")
````

## 16. Password Generator

### Sample Input:
```

```

### Sample Output:
```
nan7!k2@9#  (random)
```

````python
# Write your solution here
import random
def password(length):
  passw=""
  x="1234567890"
  y="abcdefghijklmnopqrstuvwxyz"
  a=x+y
  for i in range(length):
    passw=passw+random.choice(a)
  print(passw)
password(6)

````
